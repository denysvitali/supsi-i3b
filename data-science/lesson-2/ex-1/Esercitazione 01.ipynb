La successione di Fibonacci  una sequenza di numeri interi in cui i primi due elementi sono 1, e i successivi elementi sono dati dalla somma dei due precedenti (es. `[1,1,2,3,5,8,13,21...]`)

### 1.1
Scrivi un programma che generi una lista (che chiameremo `listFib`) contenete la successione di Fibonacci con tutti i numeri minori o uguali di $200$ (in altre parole, l'ultimo elemento della lista deve essere un numero minore o uguale di $200$).  Suggerimento: inizializza la lista con i primi due elementi `listFib = [1,1]` e poi estendila. _Da quanti elementi  composta la lista?_

````python
#Hint 
lista = [10,20,30]
print(len(lista)) # Lunghezza lista (un int)

lista.append(40)  # Aggiungere elemento ad una lista 
print(lista)

while(True):
    if len(lista) < 6:
        print("Aggiungo un elemento alla lista")
        lista.append("Nuovo elemento")
    else:
        print('Esco dal ciclo while')
        break # Esci dal ciclo while
        
print(lista) # Nota che la lista puo' contenere tipi qulsiasi

print(f"Il primo elemento della lista e' {lista[0]}")
print(f"Il terzultimo elemento della lista e' {lista[-3]}")
````

````python
# Hint: come iterare su una lista:
for elem in lista:  # Non c'e' bisogno di definire un indice i!
    print(elem)
````

````python
# Soluzione 
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)

listFib=[]
n = 0
i = 1
while n <= 200:
    listFib.append(n)
    n = fibonacci(i)
    i+=1
print(len(listFib), listFib)
````

## Esercizio 2 (dizionari)
I dizionari sono analoghi alle HashMap in Java, e funzionano nello stesso modo. [Qui](http://introtopython.org/dictionaries.html) trovate un'ottima introduzione molto graduale.  Si usano moltissimo e quindi e' importante saperli manipolare.

### Introduzione molto rapida.

````python
pop = dict() # Crea un dizionario vuoto
pop["Manno"] = 1318 # Aggiunta chiave-valore
pop["Bioggio"] = 2648
pop["Lugano"] = 63932
# Equivalente:
# pop = {"Manno" : 1318, "Bioggio" : 2648, "Lugano" : 63932}
print(pop)
````

````python
print(pop["Manno"]) # Interrogazione
print(pop.keys()) # Tutte le chiavi: agisce come una lista
print("Lugano" in pop.keys())
````

````python
print(pop.items()) # Restituisce una lista di tuple, ciascuna delle quali contiene due elementi: chiave e valore
# Per la precisione non e' una lista, ma e' un oggetto su cui si puo' iterare come se fosse una lista
````

````python
for tupla in pop.items(): # ogni elemento della simil-lista di cui sopra
    k = tupla[0]
    v = tupla[1]
    print(f"Key={k}, Value={v}")
````

````python
# Ancora meglio (usa una funzione comodissima nota come tuple unpacking)
for tupla in pop.items():
    k,v = tupla # assegna il primo e il secondo elemento di tupla a k e v rispettivamente
    print(f"Key={k}, Value={v}")
````

````python
# Ancora meglio
for k,v in pop.items():
    print(f"Key={k}, Value={v}")
````

## Esercizio 3 (stringhe)

### 3.1
Si scriva un programma Python che chieda all'utente di inserire una frase con almeno 10 parole. Per semplicit la frase non dovr contenere punteggiatura. Esempio di frase che l'utente potr inserire: "Oggi  una bella giornata Domani sar una bellissima giornata". Comando utile per inserire una stringa da linea di comando: input

Una volta inserita la frase, si scriva un programma che conti:
- il numero di parole all'interno della frase. Comando utile per creare lista dividere la stringa in parole: `split` ([documentazione](https://docs.python.org/3.6/library/stdtypes.html#str.split))
- il numero di occorrenze di ogni parola all'interno della frase. Nello specifico, si crei un dizionario dove la chiave  la parola, e il valore il numero di occorrenze della parola. Nel conteggio delle occorrenze non si faccia distinzione tra lettere maiuscole e minuscole.
- il numero di occorrenze di ogni lettera all'interno della frase. Nello specifico, si crei un dizionario simile a quello sopra. Nel conteggio delle occorrenze non si faccia distinzione tra lettere maiuscole e minuscole.
    
La cella sotto mostra come funzionano i comandi `input` e `split`

````python
# Hint
stringa = input("Inserire una frase: ")
parole = stringa.split()
print(parole)
````

````python
# Soluzione
input_str = "Oggi  una bella giornata Domani sar una bellissima giornata"
words = input_str.split(" ")

words_d = dict()
chars_d = dict()

for c in input_str:
    if c not in chars_d:
        chars_d[c] = 0
    chars_d[c] += 1

for w in words:
    if w not in words_d:
        words_d[w] = 0
    words_d[w]+=1

words_count = len(words)

print(f"{words_count} words. Words={words_d}, Chars={chars_d}")
````
