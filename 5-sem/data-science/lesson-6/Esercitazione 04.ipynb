### 1.3

Rappresentate graficamente i risultati della tabell di contingenza tramite un bar plot

```python
# Solution
# Plotto Tabella con plotly
m_mask = dfNew['sex'] == 'm'
f_mask = dfNew['sex'] == 'f'

males = dfNew[m_mask]
females = dfNew[f_mask]

smoker_by_sex = [
    [
        np.sum(smokes_mask & m_mask),
        np.sum(no_smoke_mask & m_mask)
    ],
    [
        np.sum(smokes_mask & f_mask),
        np.sum(no_smoke_mask & f_mask)
    ]
]

valori = [[],[]]
mf = ["Male", "Female"]

smokes = set(filter(lambda x: x==x, smokes)) # Filter out NaN
smokes_nosmokes_columns = ["Smokes", "Doesn't Smoke"]

for i in smokes:
    valori[0].append(np.sum(males['smokes'] == i))
for i in smokes:
    valori[1].append(np.sum(females['smokes'] == i))
t = pd.DataFrame(smoker_by_sex, index = mf, columns = smokes_nosmokes_columns)
t
```

```python
data = []
for i, name in enumerate(mf):
    data.append(go.Bar(
        x=smokes_nosmokes_columns,
        y=valori[i]/np.sum(valori[i]) * 100,
        name=mf[i]
    ))
    
    
layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Smoking"),
    yaxis=dict(title="Percentage")
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

## Esercizio 2

Si consideri il dataset ``profili.csv`` considerato nell'esercizio precedente, escludendo gli studenti. Si consideri la variabile numerica ``età`` e la variabile categorica ``Lavoratore`` che assume valori "YES" (se l'utente dichiara di avere un lavoro), "NO" (se l'utente dichiara di non avere un lavoro). La cella sotto esclude dal dataset tutti gli utenti che sono studenti o che non dichiarano il proprio lavoro e raggruppa gli utenti in base alla variabile ``Lavoratore`` definita sopra. L'array finale è ``jobArray``. Opzionale: Provate a capire i comandi eseguiti nella cella sotto.  

```python
dfNew = df.dropna(subset = ['job'])

m = len(dfNew['job'].values)
jobArray = np.full(m, "", dtype = object)

for i in range(m):
    if dfNew['job'].values[i] == 'unemployed':
        jobArray[i] = 'NO' 
    else:
        jobArray[i] = 'YES' 
```

### 2.1 
Disegnare un grouped boxplot che confronti la distribuzione dell'età per le varie sottoclassi identificate dalla variabile ``Lavoratore``.  

```python
# Solution
data = []
for val in set(jobArray):
    p_by_job = dfNew['age'][jobArray == val]
    trace = go.Box(x=p_by_job, name=val)
    data.append(trace)

layout = go.Layout(
    xaxis=dict(title="Tempo [minuti]"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### 2.2 
Simile al punto 2.1. Disegnare due istogrammi sovrapposti per confrontare la distribuzione dell'età per le varie sottoclassi identificate dalla variabile ``Lavoratore``. Perchè uno degli istogrammi non è ben visibile?

```python
# Solution
print("Uno degli istogrammi non è ben visibile perché il numero di non-lavoratori è nettamente inferiore a quello dei lavoratori")
data = []
for val in set(jobArray):
    p_by_job = dfNew['age'][jobArray == val]
    trace = go.Histogram(x=p_by_job, name=val)
    data.append(trace)

layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Età"),
    yaxis=dict(title="Persone"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### 2.3 (opzionale)

L'istogramma generato al punto precedente che rappresenta la distribuzione delle età degli utenti disoccupati non è ben visibile nella stessa figura dell'istogramma che rappresenta la distribuzione delle età degli utenti lavoratori poichè il numero di utenti disoccupati è molto piu' basso rispetto ai lavoratori. Di conseguenza, anche le barre dell'istogramma sono molto piu' basse rispetto alle barre dell'istogramma caratterizzante gli utenti lavoratori. La cella sotto mostra come generare un istogramma normalizzato (basta impostare l'attributo ``histnorm`` uguale a ``probability'`` quando si creare l'oggetto della classe ``go.Histogram``).  

```python
#Hint
x = np.random.randn(1000)

trace = go.Histogram(x = x, histnorm = 'probability', marker = dict(color = 'rgb(255,0,0)'))
layout = go.Layout(title = 'Esempio di istogramma normalizzato', xaxis = dict(title = 'x'), yaxis = dict(title = 'frequenze relative'))
fig = go.Figure([trace], layout)
py.iplot(fig)
```
